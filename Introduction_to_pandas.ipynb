Pandas is a python package for handling tabular data. It is the python equivalent of excel, calc or simmilar.

```python
import pandas as pd # first import the package
```

## Loading / generating data
You can either import data from a file or generate it yourself directly here in the note book.
<br>We start with the import of text based files (.csv, .tsv,...). The function `read_csv()` is all you need.
<br>If you want to oepn a .csv (comma separated values) file the default works fine, for for other text file formats you must add the correct delimiter.

As example for a .tsv (tab separated values) file you must provide the string '\t'.

```python
df = pd.read_csv('data/inflammation-01.csv')
# for a .tsv file: pd.read_csv('path to a file', delimiter='\t')


display(df) # display shows our table in a nice way (works only in jupyter!!!)
```

If we add a `.head()` behind the variable containing our tabular data only the first five rows will be displayed. The equivalent to display the last five rows is `.tail()`.

```python
display(df.head())
display(df.tail())
```

We loaded the file but the first rwo of data was taken as column headers.
<br>To add the correct headers we must first define a list containing matching column names.

```python
columns = []
for i in range(1, 41):
    columns.append(f'Day{i}')

print(columns)
```

To load the file with our custom column names we must provide them through the names argument.

```python
df_with_headers = pd.read_csv('data/inflammation-01.csv', names=columns)
display(df_with_headers.head())
```

The moment we loaded our data from the file into a variable we created an object called a pandas `DataFrame`. This is just a container for our data with many usefull functions. 
<br>Similar to a numpy array we can ask it for its shape.

```python
df_with_headers.shape
```

Our dataframe has the correct number of rows and columns.
<br>Let's add a meaningfull index to the table.
<br>First we have to create the index. Therefore, we first create a list containing all the row IDs in our case the patient numbers.

```python
index = []
for i in range(1,61):
    index.append(f'Patient{i}')
print(index)
```

In the next step we must change the list to an actual pandas `Index` object and then add it to our dataframe.

```python
real_index = pd.Index(index) # here we change our list index to an index object
df_with_index_and_headers = df_with_headers.set_index(real_index) # we set the index of our dataframe
display(df_with_index_and_headers.head())
```

Previously, we loaded data different example files and added meaningful column names and row indices, let's try to make an entirely new table next.
<br>Therefore, we use the pandas `DataFrame` class directly and fill it from a python dictionary (`dict`). A dictionary works similar to a real life dictionary, 
<br>you have keys you can look for and values that are affiliated with a key: 'animal': 'Tier'.

```python
mytable = pd.DataFrame(dict(
    col1=[1,2,3], # this is our first column with three rows
    col2=['a','b','c'] # this is our second column
))
display(mytable)
```

It is important that each column has the same number of elements (rows), otherwise ou will run into an error.

```python
mytable = pd.DataFrame(dict(
    col1=[1,2,3,4,5],
    col2=['a','b','c']
))
display(mytable)
```

## Navigating a table
We can select single or multiple columns of our dataframe by their names.

```python
df_from_excel['FamilyName'] # select a single column
```

```python
df_from_excel[['FamilyName','GivenName']] # select multiple columns (care to not forget the double brackets)
```

When we select multiple columns the order of the column names defines the oder the columns are displayed.

```python
df_from_excel[['GivenName', 'FamilyName']]
```

We can as well select a single or multiple rows of our dataframe. Therefore, we have to use `.loc`.

```python
df_from_excel.loc[2] # single row
```

```python
df_from_excel.loc[1:2] # multiple rows, mind that the upper und lower bounds are bouth included
```

It is important to use the actual indices of our table! If we modify the index to custom names we have to use the custom names

```python
df_with_index_and_headers.loc['Patient1']
```

We can use the row and columns selection together to select a specific subset of our table or a specific cell.

```python
df_from_excel.loc[2, 'FamilyName'] # specific cell
```

```python
df_from_excel.loc[1:2, ['FamilyName', 'GivenName']]
```

## Modifying a table
We can modify dataframes in many ways using pandas. I will just give two simple examples, for more take a look at the pandas documantation (https://pandas.pydata.org/docs/index.html) please.
<br>The first example will be to add a new column. To add a new column we can simply tell the dataframe the name of the column and its content.

```python
display(df_from_excel)
```

```python
df_from_excel['FaveColor'] = ['green', 'yellow', 'blue', 'orange']
display(df_from_excel)
```

The second example is to combine two dataframes. We can use the `concat()` function to combine two dataframes in various ways.
<br>Here we will just append. First we create a new dataframe with additional persons (just for this example), afterwards we combine the two dataframes. 

```python
new_df = pd.DataFrame(dict(
    FamilyName=['Turner','Bently'],
    GivenName=['Hanna','Joe'],
    Age=[5, 83],
    Size=[1.14, 1.67],
    Weight=[21, 82],
)) # we do not add their fave color for a specific reason, just take a look at the table below

combined_df = pd.concat([df_from_excel, new_df]) # the brackets are important
display(combined_df)
```

We did not add the fave color to our new dataframe, because we did not need to. `concat()` can combine dataframes even if columns are missing in one or both of the dataframes.
<br>Notice that the index is made from both dataframes. If the values of the indices do not have any special meaning you can reset them with the argument `ignore_index=True`.

```python
combined_df = pd.concat([df_from_excel, new_df], ignore_index=True)
display(combined_df)
```

## Saving a table
The last step when handling dataframes might be to save them back to a file you can store somewhere or give your collegues.
<br>To save da dataframe just call the `.to_csv()` method to save as a text file or the `.to_excel()` method to save as an excel file.
<br>If your dataframe does not have an index of any meaning we do not save it by adding the argument: `index=False`.

```python
df_from_excel.to_csv('data/my_fancy_table.csv', index=False) # save to a .csv file, ignore the index 
```

```python
df_from_excel.to_excel('my_fancy_table.xlsx', index=False) # save to an excel file, ignore the index
```
