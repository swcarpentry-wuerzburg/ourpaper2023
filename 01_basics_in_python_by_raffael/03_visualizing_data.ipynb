# Importing data with numpy

We want to visualize our patient data in plots. For that we load our data like before.

```python
# import numpy for data read/write and manipulation
import numpy as np
data = np.loadtxt(fname='../data/inflammation-01.csv', delimiter=',')
```

To visualize data, we need to also import the package `matplotlib.pyplot`, which we will abbreviate by `plt`.

```python
# import matplotlib.pyplot for data visualization
import matplotlib.pyplot as plt

# visualize the raw data using imshow() 
plt.imshow(data)
plt.show()
```

# Creating a basic plot: plotting many curves in one figure

We create a basic plot now. First, we use the function `plot()` from the package `matplotlib.pyplot` to set up our first graph. By not specifying the x-values of the plot, it simply plots the index of our data set.

# saving a plot to a file

Save the plot above with the function `savefig()`. We have to give the function a filename as a string, here we use `inflammation.*`, where `*` tells the program in which filetype I want to store the figure.

Note that `savefig()` should be the last command before `show()` to save all modifications to the figure.

```python
# The structure is always: plt.plot( <data y-values> , <other options> ), where other options are modifying a curve.
plt.plot( np.mean(data,axis=0) , color = 'blue',  linestyle = 'dotted', label = 'average', linewidth = 1)
plt.plot( np.max(data,axis=0) , color = 'green', linestyle = 'dashed', label = 'maximum', linewidth = 2 )
plt.plot( np.min(data,axis=0) , color = 'black', linestyle = 'solid', label = 'minimum', linewidth = 3)

# set x- and y- labels and also the title of the plot
plt.xlabel('day')
plt.ylabel('inflammation')
plt.title('inflammation each day')

# show the legend in the plot, with labels we specified above in the plot() function
plt.legend()

# create a pixel image by replacing * with png
plt.savefig('inflammation.png')

# create a pixel image with a higher resolution by specifying the dots per inch (dpi)
plt.savefig('inflammation_high_res.png',dpi=600)

# create a vector graphics by replacing * with pdf
plt.savefig('inflammation.pdf')

# create a vector graphics that can be edited with other programs: use svg as the file ending
plt.savefig('inflammation.svg')

# show the plot below
plt.show()
```

# Grouping: plotting each curve in separate panel in a larger figure

```python
fig = plt.figure(figsize=(10.0,3.0))

axes1 = fig.add_subplot(1,3,1)
axes2 = fig.add_subplot(1,3,2)
axes3 = fig.add_subplot(1,3,3)

axes1.set_ylabel('average')
axes2.set_ylabel('maximum')
axes3.set_ylabel('minimum')

axes1.plot( np.mean(data,axis=0) , linestyle = 'dashed' , color = 'blue', marker = 'o')
axes2.plot( np.max(data,axis=0) , linestyle = 'solid' , color = 'green', marker = '^')
axes3.plot( np.min(data,axis=0) , linestyle = 'dotted' , color = 'black', marker = 'd', markersize = 5)

axes1.set_xlabel('day')
axes2.set_xlabel('day')
axes3.set_xlabel('day')

fig.tight_layout()

#plt.savefig('inflammation.png')
#plt.savefig('inflammation2.png', dpi = 300)
#plt.savefig('inflammation.pdf')
plt.show()
```
