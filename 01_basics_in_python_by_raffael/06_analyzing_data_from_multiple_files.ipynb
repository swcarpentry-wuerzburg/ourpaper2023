# Accessing many data files at once

```python
# the library glob has a function glob() that finds files and directories whose names match a pattern
import glob

# get the path of the data files that have 'inflammation' and 'csv' in their name
# sorted() sorts the files in ascending order
filenames = sorted( glob.glob('../data/inflammation*.csv') )
print(filenames)
```

```python
# we get a list of strings that contains the pathes to our files
# for instance, element 2 in the list is
filenames[2]

# we can use these strings to load many files in a loop
```

```python
import numpy as np
import matplotlib.pyplot as plt

#filenames = filenames[0:3]
for filename in filenames:
    print(filename)

    data = np.loadtxt(fname=filename, delimiter=',')

    fig = plt.figure(figsize=(10.0, 3.0))

    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)

    axes1.set_xlabel('day')
    axes1.set_ylabel('average')
    axes1.plot(np.mean(data, axis=0))

    axes2.set_xlabel('day')
    axes2.set_ylabel('max')
    axes2.plot(np.max(data, axis=0))

    axes3.set_xlabel('day')
    axes3.set_ylabel('min')
    axes3.plot(np.min(data, axis=0))

    fig.tight_layout()
    plt.show()
```

The second set of data looks similar to the first one. The third looks very differently. There might be something wrong with the set of data or how it was collected. Let's look at a heatmap to further investigate this set.

```python
data = np.loadtxt(fname=filenames[2], delimiter=',')
plt.imshow(data)
plt.show()
```
